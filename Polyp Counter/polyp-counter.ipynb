```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "../input/kvasir-polyp-testset/MedAI_2021_Polyp_Segmentation_Test_Dataset/"
```

```python
mask_path = "../input/polyptestsetpredresults/"
```

```python
for i in os.listdir("../input/polyptestsetpredresults"):
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    mask = cv2.imread(mask_path + i, cv2.IMREAD_GRAYSCALE )
    plt.imshow(mask)
    th, threshed = cv2.threshold(mask, 100, 255,cv2.THRESH_BINARY_INV|cv2.THRESH_OTSU)
    contours,hierarchy = cv2.findContours(threshed,1,2)
    plt.title("Number of polyps: {}".format(len(contours)-1), fontsize=20)
    plt.subplot(1,2,2)
    img = cv2.imread(img_path + i, cv2.IMREAD_COLOR)
    plt.imshow(img)
    
    plt.show()
```
